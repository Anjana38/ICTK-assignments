## Whether Analyzer

```python
# Sample weather data for a region over a week (temperature in °C, hunidity in %)
temperatures = [32, 29, 30, 33, 28, 32, 35]
humidities = [81, 80, 76, 75, 76, 74, 78]


#function to calculate average temperature
def average_temperature(data):
  return sum(data)/len(data)

# Function to calculate average humidity
def average_humidity(data):
  return sum(data)/len(data)

# Function to print whether summary
def weather_summary():
  avg_temp = average_temperature(temperatures)
  avg_hum = average_humidity(humidities)
  print("Weather Summary of Kochi in this week")
  print(f"The average temperature is {avg_temp:.2f}°C")
  print(f"The average humidity is {avg_hum:.2f}%")

  # Conditions
  if avg_temp > 32 and avg_hum > 70 :
    print("Hot and humid week")
  elif avg_temp < 32 and avg_hum > 75:
    print("Rainy week")
  else:
    print("Moderate climate")

weather_summary()

```
