##  10. Finding Anomalies

Identify potential outliers in the mpg column using the Interquartile Range (IQR)
method. Specifically:


● Calculate the IQR for mpg.

● Define outliers as values less than Q1 - 1.5 * IQR or greater than Q3 +
1.5 * IQR.


● Create a DataFrame of cars classified as outliers, displaying car_name, mpg,
and model_year.

```python
Q1 = data['mpg'].quantile(0.25)
Q3 = data['mpg'].quantile(0.75)
IQR = Q3 - Q1
print(f'\n lower Quartile is',Q1)
print(f'\n upper Quartle is',Q3)
print('\n IQR is',IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f'\n lower bound is',lower_bound)
print(f'\n upper bound is',upper_bound)
```

```python
outlier_data = data[(data['mpg'] < lower_bound) | (data['mpg']> upper_bound)][['car name','mpg','model year']]
```

```python
outlier_data
```

## 9. Correlation
Compute the Pearson correlation coefficient between mpg and weight using
NumPy.

```python
data = np.genfromtxt('auto-mpg.csv', delimiter=',', skip_header=1)
```

```python
mpg = data[:,0]
weight = data[:,4]
```

```python
correlation_matrix = np.corrcoef(mpg,weight) # np.corrcoef() returns correlation matrix
```

```python
pearson_coef = correlation_matrix[0, 1]
```

```python
print ('Pearson coefficient is ', pearson_coef)
```

## 10.  Conditional Aggregates
Calculate the mean mpg for cars grouped by the number of cylinders using NumPy
techniques.

```python
unique_cylinders = np.unique(cylinders_array)
unique_cylinders
```

```python
for c in unique_cylinders:
  mpg_mean = np.mean(mpg_array[cylinders_array == c])
  print(f'Cylinders :{c}, mean mpg is {mpg_mean}')
```
